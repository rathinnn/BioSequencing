```python
def stringgapped(path,k,d):
    ret=''
    pre=[]
    suf=[]
    prefix=''
    suffix=''
    for pair in path:
        temp=pair.split("|")
        pre.append(temp[0])
        suf.append(temp[1])
    for i in range(len(pre)):
        if(i==len(pre)-1):
            prefix=prefix+pre[i]
        else:
            temp1=pre[i]
            prefix=prefix+temp1[0]
    for i in range(len(suf)):
        if(i==len(suf)-1):
            suffix=suffix+suf[i]
        else:
            temp1=suf[i]
            suffix=suffix+temp1[0]
    ret=prefix+suffix[len(suffix)-k-d:]
    
    return ret
```
